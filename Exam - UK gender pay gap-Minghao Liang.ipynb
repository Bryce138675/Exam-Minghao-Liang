What we are going to do here is fit a probability density function (PDF) to the data. A PDF is a statistical model of our data that we can using some prior assumption about how the data is distributed.

First fun this cell for a function for calculating a pdf.

```python
def make_pdf(dist, params, size=10000):
    """Generate distributions's Probability Distribution Function """

    # Separate parts of parameters
    arg = params[:-2]
    loc = params[-2]
    scale = params[-1]

    # Get sane start and end points of distribution
    start = dist.ppf(0.01, *arg, loc=loc, scale=scale) if arg else dist.ppf(0.01, loc=loc, scale=scale)
    end = dist.ppf(0.99, *arg, loc=loc, scale=scale) if arg else dist.ppf(0.99, loc=loc, scale=scale)

    # Build PDF and turn into pandas Series
    x = np.linspace(start, end, size)
    y = dist.pdf(x, loc=loc, scale=scale, *arg)
    pdf = pd.Series(y, x)

    return pdf
```

Lets try fitting a PDF using a Gaussian distribution.

```python
dist_f = st.norm
params = dist_f.fit(Male_bottom_quartile_data)
pdf = make_pdf(dist_f, params)

plt.figure(figsize=(8,6))
ax = pdf.plot(lw=2, label='PDF', legend=True)
Male_bottom_quartile_data.plot(kind='hist', bins=50, density=True, alpha=0.5, label='Data', legend=True, ax=ax)
ax.set_ylabel('Density')
```

Lets try a different distribution this time, lets try the exponential distribution:

```python
dist_f = st.expon
params = dist_f.fit(Male_bottom_quartile_data)
pdf = make_pdf(dist_f, params)

plt.figure(figsize=(8,6))
ax = pdf.plot(lw=2, label='PDF', legend=True)
Male_bottom_quartile_data.plot(kind='hist', bins=50, density=True, alpha=0.5, label='Data', legend=True, ax=ax)
ax.set_ylabel('Density')
```

This time lets try a skewed normal distribution:

```python
dist_f = st.skewnorm
params = dist_f.fit(Male_bottom_quartile_data)
pdf = make_pdf(dist_f, params)

plt.figure(figsize=(8,6))
ax = pdf.plot(lw=2, label='PDF', legend=True)
Male_bottom_quartile_data.plot(kind='hist', bins=50, density=True, alpha=0.5, label='Data', legend=True, ax=ax)
ax.set_ylabel('Density')
```

Let's try a lognormal distribution instead:

```python
dist_f = st.lognorm
params = dist_f.fit(Male_bottom_quartile_data)
pdf = make_pdf(dist_f, params)

plt.figure(figsize=(8,6))
ax = pdf.plot(lw=2, label='PDF', legend=True)
Male_bottom_quartile_data.plot(kind='hist', bins=50, density=True, alpha=0.5, label='Data', legend=True, ax=ax)
ax.set_ylabel('Density')
```

# Calculate BIC scores

```python
from sklearn.mixture import GaussianMixture
from sklearn.model_selection import GridSearchCV
```

```python
def gmm_bic_score(estimator, X):
    """Callable to pass to GridSearchCV that will use the BIC score."""
    # Make it negative since GridSearchCV expects a score to maximize
    return -estimator.bic(X)


param_grid = {
    "n_components": range(1, 7),
    "covariance_type": ["spherical", "tied", "diag", "full"],
}
grid_search = GridSearchCV(
    GaussianMixture(), param_grid=param_grid, scoring=gmm_bic_score
)
grid_search.fit(x_scaled)
```

```python
bic_scores = pd.DataFrame(grid_search.cv_results_)[
    ["param_n_components", "param_covariance_type", "mean_test_score"]
]
bic_scores["mean_test_score"] = -bic_scores["mean_test_score"]
bic_scores = bic_scores.rename(
    columns={
        "param_n_components": "Number of components",
        "param_covariance_type": "Type of covariance",
        "mean_test_score": "BIC score",
    }
)
bic_score = bic_scores.sort_values(by="BIC score").head()
print(bic_score)
```

# Can you model you use one or more variables to predict another, or demonstrate a relationship between two or more variables?

# Making Models
So we have determined that we think there is a linear relationship between these two variables, now we'll look at how we can build a linear model to fit this relationship.

Having a model allows us to make predictions about new valuese

Formula of a line
We can define the function for any line in the form

y = ax + b + e

Where y is the dependent variable (surveillance), x is the independent variable (population), a and b are two parameters for us to set and e is the error.

# .stats.linregress()
And we can do that using linregress() from the scipy.stats package. This returns us a value for the slope (a) and intercept (b).

We can then plot the line, looks like a pretty good model!

```python
data = df3[["Male bottom quartile %","Female bottom quartile %"]].values
x = data[:,0]
y = data[:,1]
slope, intercept, r, p, std_err = stats.linregress(x, y)
slope = np.float32(slope)
model_y = [slope * i + intercept for i in x]

ax = plot_scatter(x,y,"Male bottom quartile %","Female bottom quartile %")
ax.plot(x, model_y)
```

Female bottom quartile %= -Male bottom quartile %+100

With an intercept of 8.5 and a slope of 1.85, Female bottom quartile %+Male bottom quartile %=100

I could use the Male bottom quartile % to predict  Female bottom quartile %.

# Evaluate
r^2

We use the r^2 (r squared), otherwise known as the coefficient of determination to evaluate the fit of our model. This tells us how much of the variation in the dependent variable (y) is explained by the model.

An r^2 of 1 is a perfect model and an r^2 of 0 means the model is no better than just picking the mean for every prediction.

```python
r
```

```python
r**2
```

This is a perfect model.

# p-value
We also get a p-value returned by our model, and we know that this is result of a hypothesis test.

The hypothesis we are looking to test is that there is a non-zero slope for this model, making the null hypothesis that the slope is 0. A p-value of less than 0.05 allows us the reject the null hypothesis and conclude there is a relationship between the two variables



```python
p
```

# Multiple Regression
Above, we have seen that sometimes just using one independent variable to predict another fails to capture the whole relationship. In this case we can build a model that takes multiple variables into account.

For example, if we had 3 values for each example, we could try and fit the model below

y = ax1 + bx2 + cx3 + d + error

```python
from sklearn import linear_model 
from sklearn.metrics import mean_squared_error, r2_score
gendar_df = pd.read_csv("data/UK-Gender-Pay-Gap-2018-19.csv") 
x = gendar_df[['Male lower middle quartile %','Male bottom quartile %']]
y = gendar_df['Male upper middle quartile %'] 
regr = linear_model.LinearRegression()
regr.fit(x, y) 
print(regr.coef_, regr.intercept_,regr.score(x, y)) 
```
